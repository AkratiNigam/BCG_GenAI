IMPORTING LIBRARIES

```python
import pandas as pd
```

READING THE DATA

```python
data = pd.read_csv("BCG TASK 1.csv")
data
```

```python
data.isna()
```

```python
data = data.sort_values(['Company','Year'],ascending=True).groupby('Company').head()
data
```

ANALYZING EACH COMPANY'S GROWTH

% OF REVENUE GROWTH

```python
data_revenue = pd.DataFrame(columns=['Company','Year','Total Revenue','% of Revenue Growth'])
data_revenue['Company'] = data['Company']
data_revenue['Year'] = data['Year']
data_revenue['Total Revenue'] = data['Total Revenue']
data_revenue['% of Revenue Growth'] = data.groupby(['Company'])['Total Revenue'].pct_change() * 100
data_revenue
```

APPLE:  After a revenue dip in 2023 (-2.80% growth), Apple bounced back with 2.02% growth in 2024.
MICROSOFT:  Consistent growth each year, with a remarkable 15.67% revenue growth in 2024.
TESLA:  Strong growth in 2023 (18.79%), followed by minimal growth in 2024 (0.95%).
Microsoft showed the strongest upward trend in revenue growth over the three years, while Tesla's growth slowed in 2024. Apple rebounded after a decline in 2023. 



% OF NET INCOME GROWTH

```python
data_income = pd.DataFrame(columns=['Company','Year','Net Income','% of Net Income Growth'])
data_income['Company'] = data['Company']
data_income['Year'] = data['Year']
data_income['Net Income'] = data['Net Income']
data_income['% of Net Income Growth'] = data.groupby(['Company'])['Net Income'].pct_change() * 100
data_income
```

APPLE: A decrease of 2.81% in 2023 followed by a 3.36% decrease in 2024 signals ongoing challenges.
MICROSOFT: A significant rebound in 2024, with net income growth at an impressive 1118.00% after a decline in 2023 (-90.05%).
TESLA: Positive net income growth in 2023 (18.96%), but a sharp decline of -52.23% in 2024 indicates potential volatility.
Microsoft shows strong recovery, while Tesla experiences a dramatic shift in 2024. Apple, meanwhile, faces consistent declines.

% OF ASSETS GROWTH

```python
data_assets = pd.DataFrame(columns=['Company','Year','Total Assets','% of Assets Growth (%)'])
data_assets['Company'] = data['Company']
data_assets['Year'] = data['Year']
data_assets['Total Assets'] = data['Total Assets']
data_assets['% of Assets Growth'] = data.groupby(['Company'])['Total Assets'].pct_change() * 100
data_assets
```

% OF LIABILITIES GROWTH

```python
data_liability = pd.DataFrame(columns=['Company','Year','Total Liabilities','% of Liabilities Growth'])
data_liability['Company'] = data['Company']
data_liability['Year'] = data['Year']
data_liability['Total Liabilities'] = data['Total Liabilities']
data_liability['% of Liabilities Growth'] = data.groupby(['Company'])['Total Liabilities'].pct_change() * 100
data_liability
```

APPLE: Liabilities declined in 2023 (-3.86%) but grew by 6.06% in 2024.
MICROSOFT: Consistent growth in liabilities, with a modest increase of 3.76% in 2023, followed by significant growth of 18.44% in 2024.
TESLA: Strong liabilities growth in 2023 (18.03%) and continued but slower growth in 2024 (12.51%).
Microsoft leads in liabilities growth for 2024, while Tesla demonstrates consistent but tapering increases. Apple shows a recovery trend after a decrease.

% OF CASHFLOW

```python
data_cashflow = pd.DataFrame(columns=['Company','Year','Cashflow','% of Cashflow Growth'])
data_cashflow['Company'] = data['Company']
data_cashflow['Year'] = data['Year']
data_cashflow['Cashflow'] = data['Cashflow']
data_cashflow['% of Cashflow Growth'] = data.groupby(['Company'])['Cashflow'].pct_change() * 100
data_cashflow
```

APPLE: After a significant decline in 2023 (-9.50%), Apple recovered with a 6.98% increase in cash flow for 2024.
MICROSOFT: Minor decline in 2023 (-1.63%), followed by an impressive growth of 35.36% in 2024.
TESLA: Sharp drop in 2023 (-9.97%), but a strong rebound in 2024 with a 12.64% increase in cash flow.
Microsoft demonstrated the most remarkable recovery and growth in 2024, while Tesla and Apple showed positive trends after declines in 2023.
